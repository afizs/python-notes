### Nested Functions

```python
def foo(): # outer function
    def say(): # inner function
        print('Hello World!')
    return say
```

```python
a = foo()
```

```python
a()
```

### Criteria for closures
1. There should be nested functions
2. The inner function should be using the variable of outer function
3. Outer function should return the inner function

### When to use Closures?
Closures can avoid the use of global values and provides some of form of data abstraction. 

### Example of Clossures in Python 

```python
def multiplier(x): # outer function and x is free variable here
    def multiply(y):
        return x * y
    return multiply
```

```python
m_2 = multiplier(2)
m_3 = multiplier(3)
m_9 = multiplier(9)
```

```python
m_2(6)
```

```python
m_3(7)
```

```python
m_9(8)
```

### Example 2: Counter of the function calls

```python
def hello():
    count = 0
    def say(name):
        nonlocal count
        count = count + 1
        print(f'calling func times: {count}')
        return f'Hello {name}'
    return say
```

```python
a = hello()
a('Afiz')
```

```python
a('Rayan')
```

```python
b = hello()
```

```python
b('Neelu')
```
