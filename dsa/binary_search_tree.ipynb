```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left, self.right = None, None
```

#### Insert 

```python
def insert(node, key):
    if node is None:
        node = Node(key)
    elif node.key > key:
        node.left = insert(node.left, key)
    elif node.key < key:
        node.right = insert(node.right, key)
    return node
```

#### Inorder Traversal

```python
def inorder(node):
    if node is None:
        return []
    return (inorder(node.left) + [node.key] + inorder(node.right))
```

```python
r = Node(50)
for i in [20, 60, 10, 70, 5, 25, 65, 55]:
    r = insert(r, i)
```

```python
inorder(r)
```

# âœ… Traversing a Binary Tree

### Inorder Tree Traversal

```python
def inorder(node):
    if node is None:
        return []
    return (inorder(node.left) + [node.key] + inorder(node.right))

inorder(r)
```

#### Pre-Order Tree traversal

```python
def preorder(node):
    if node is None:
        return []
    return ([node.key]+ preorder(node.left) + preorder(node.right))
preorder(r)
```

#### Post-Order Tree traversal

```python
def postorder(node):
    if node is None:
        return []
    return (postorder(node.left) +  postorder(node.right) + [node.key])
postorder(r)
```

```python
def height_of_bst(node):
    if node is None:
        return 0
    return 1 + max(height_of_bst(node.left), height_of_bst(node.right))
```

```python
def size_of_bst(node):
    if node is None:
        return 0
    return 1 + size_of_bst(node.left) + size_of_bst(node.right)
```
